```python
from nltk import sent_tokenize, word_tokenize
import itertools
from razdel import sentenize
from pymorphy2 import MorphAnalyzer
from collections import Counter, defaultdict
import numpy as np
import re
from string import punctuation
from nltk.corpus import stopwords
import json
from math import log
import nltk
nltk.download('stopwords')
```

```python
stops = set(stopwords.words('russian') + ["это", "весь"])
morph = MorphAnalyzer()
```

```python
def normalize(text):
    tokens = re.findall('[а-яёa-z0-9]+', text.lower())
    normalized_text = [morph.parse(word)[0].normal_form for word in tokens]
    normalized_text = [word for word in normalized_text if len(word) > 2 and word not in stops]
    
    return normalized_text
```

```python
def preprocess(text):
    sents = sentenize(text)
    return [normalize(sent.text) for sent in sents]
```

```python
def ngrammer(tokens, stops, n=2):
    ngrams = []
    tokens = [token for token in tokens if token not in stops]
    for i in range(0,len(tokens)-n+1):
        ngrams.append(tuple(tokens[i:i+n]))
    return ngrams
```

```python
twitter['text_norm'] = twitter.comment.progress_apply(normalize)
```

```python
text_transformer = TfidfVectorizer(ngram_range=(1, 2), max_features=15000)
```

```python
lenta['data'] = 'lenta'
twitter['data'] = 'twitter'
```

```python
df = pd.concat([lenta, twitter])[['text_norm','data']]
```

```python
df.rename(columns={'text_norm':'text'}, inplace=True)
```

```python
df.head()
```

```python
df.text = df.text.apply(lambda x: ' '.join(x))
```

```python
text_transformer.fit(df.text)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df)
```

```python
X_train_text = text_transformer.fit_transform(train['text'])
X_test_text = text_transformer.transform(test['text'])
```

```python
X_train_text.shape, X_test_text.shape
```

```python
logit = LogisticRegression(C=5e1, solver='lbfgs', multi_class='multinomial', random_state=17, n_jobs=4)
```

```python
logit.fit(X_train_text, train['data'])
```

```python
eli5.show_weights(estimator=logit, 
                  feature_names = list(text_transformer.get_feature_names()),
                 top=(50, 5))
```
